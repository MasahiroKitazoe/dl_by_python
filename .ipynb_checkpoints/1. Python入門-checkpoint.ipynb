## Numpyのブロードキャスト

```python
import numpy as np
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([10, 20])
A * B
```

- １次元配列のBが変形されて2次元配列のAと同じようになって、演算が行われている

## 要素へのアクセス

```python
X = np.array([[51, 55], [14, 19], [0, 4]])
print(X)
```

```python
X[0] #0行目
```

```python
X[0][1] #(0, 1)の要素
```

```python
for row in X:
    print(row)
```

```python
#上記のインデックス操作に加えて、配列によるアクセスも可能
X = X.flatten()
print(X)
```

```python
X[np.array([0, 2, 4])]
```

- Numpyの配列に対して不等号などの演算子を使うと、結果はbooleanの配列になる
- この配列を使って、配列の各要素を取り出している

```python
X > 15
```

```python
X[X>15]
```

## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 6, 0.1) #0から6まで0.1刻みで生成
y = np.sin(x)
```

```python
# グラフの描画
plt.plot(x, y)
plt.show()
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
#グラフの描画
plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle="--", label='cos') #破線で描画
plt.xlabel('x')# x軸のラベル
plt.xlabel('y')# y軸のラベル
plt.title('sin & cos')#タイトル
plt.legend()
plt.show()
```
